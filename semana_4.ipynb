Una empresa desea proteger la información que se envía entre sus usuarios mediante un sistema simple de cifrado. Para ello, cada letra del alfabeto se reemplaza por otra letra según una tabla de correspondencia predefinida.

Desarrolle un programa que solicite un mensaje y genere su versión cifrada aplicando la siguiente tabla de sustitución:

| Original | a | b | c | d | e | f | g | h | i | j | k | l | m | n | ñ | o | p | q | r | s | t | u | v | w | x | y | z |
| --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- | --- |
| Cifrada | n | f | l | q | t | k | o | u | r | c | y | a | v | j | d | h | w | e | i | ñ | m | x | g | p | z | b | s |

Para realizar la sustitución, utilice las siguientes cadenas:

```python
abecedario = "abcdefghijklmnopqrstuvwxyz"
cifrado = "nflqtkourcyavjdhweimxgpzbs"
```

**Consideraciones:**

- Los mensajes de entrada contendrán únicamente letras del alfabeto inglés (`a-z`) y espacios.
- Los espacios no forman parte de la tabla de sustitución y deben conservarse sin modificaciones en el mensaje cifrado.
- No modifique el contenido de las variables `abecedario` y `cifrado`.

**Ejemplo de ejecución:**

```
Ingrese el mensaje a cifrar: Hola mundo
Mensaje cifrado: udan vxjqd
```

##
**Solución Propuesta:**

````python
# Alfabeto normal
abecedario = "abcdefghijklmnopqrstuvwxyz"

# Alfabeto cifrado
cifrado = "nflqtkourcyavjdhweimxgpzbs"

# Convierte todo el mensaje a minúsculas para evitar sensibilidad a mayúsculas
mensaje = input("Ingrese el mensaje a cifrar: ").lower()

# Variable donde se almacenará el mensaje cifrado
resultado = ""

# Recorre cada carácter del mensaje ingresado
for letra in mensaje:

    # Si el carácter es un espacio, se conserva
    if letra == " ":
        resultado += " "

    # Si es una letra, se busca su posición en el abecedario
    else:
        indice = abecedario.index(letra)
        # Se reemplaza por el carácter correspondiente del cifrado
        resultado += cifrado[indice]

print("Mensaje cifrado:", resultado)
````

## Ejercicio 1

En un videojuego existen 12 cofres misteriosos. Algunos cofres contienen recompensas y otros piedras sin valor.

Dada la lista de premios:

- `premios = ["Diamantes", "Perla mágica", "Monedas de oro", "Sombrero de capitán", "Espada pirata", "Mapa del tesoro"]`

Cree un programa que retorne una lista de 12 elementos simulando estos cofres. En esta lista:

- 4 cofres aleatorios deben contener premios aleatorios seleccionados de la lista de premios. Los premios pueden repetirse.
- El resto de los cofres deben contener `"Piedra"`

Además, haga que el programa permita al usuario elegir un cofre para abrir.

- El usuario tiene la oportunidad de abrir 3 cofres y obtener premios.
- Si el cofre contiene un premio, se debe mostrar el nombre del premio. `"¡Felicidades! Has ganado: recompensa"`
- Si contiene una piedra, se debe mostrar el mensaje `"¡Arrr! Solo había una piedra en este cofre"`
- Si el usuario ingresa cualquier otra cosa, se debe mostrar el mensaje `"Por favor, ingrese un número entre 1 y 12"`
- Cuando el juego finalice, muestre los premios ganados por el jugador.

**Pista:** Para la creación de la lista sorteada no es necesario del uso de bucles. Utilice la función choices()

**Ejemplo de ejecución:**

```
¡Bienvenido al juego de los cofres del tesoro!

Intento 1 - Elija un cofre para abrir (1-12): 100
Por favor, ingrese un número entre 1 y 12

Intento 2 - Elija un cofre para abrir (1-12): 10
¡Felicidades! Has ganado: Sombrero de capitán

Intento 3 - Elija un cofre para abrir (1-12): 5
¡Arrr! Solo había una piedra en este cofre

Juego terminado. Premios ganados: ['Sombrero de capitán']
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 1 aquí

````

##
**Solución Propuesta:**

````python
import random as rd

premios = ["Diamantes",
           "Perla mágica",
           "Monedas de oro",
           "Sombrero de capitán",
           "Espada pirata",
           "Mapa del tesoro"
           ]

# Creación de la lista de premios y piedras

# Selecciona 4 premios al azar de la lista, se pueden repetir
aleatorios = rd.choices(premios, k=4)
# Crea una lista con 8 piedras
piedras = ["Piedra"]*8
# Une los premios y las piedras en una sola lista de 12 elementos
cofres = aleatorios + piedras
# Mezcla los elementos para que los premios queden en posiciones aleatorias
rd.shuffle(cofres)

# Lista donde se guardarán los premios obtenidos por el jugador
premios_ganados = []

print("¡Bienvenido al juego de los cofres del tesoro!\n")

# El jugador tiene 3 oportunidades para abrir cofres
for intento in range(3):
    seleccion = input(f"Intento {intento + 1} - Elija un cofre para abrir (1-12): ")

    # Verifica que la entrada sea un número entre 1 y 12
    if seleccion.isdigit() and 1 <= int(seleccion) <= 12:

        # Convierte el número ingresado a índice de lista
        indice = int(seleccion) - 1

        # Comprueba el contenido del cofre seleccionado
        if cofres[indice] == "Piedra":
            print("¡Arrr! Solo había una piedra en este cofre\n")
        else:
            print(f"¡Felicidades! Has ganado: {cofres[indice]}\n")
            premios_ganados.append(cofres[indice])

    # Si la entrada no es válida, muestra un mensaje de error
    else:
        print("Por favor, ingrese un número entre 1 y 12\n")

print("Juego terminado. Premios ganados:", premios_ganados)
````

Los URLs normalmente se repiten y corresponden algunas veces a universidades de Ecuador y otros países. Note que los URLs no diferencian entre mayúsculas y minúsculas. Por ejemplo: `www.espol.edu.ec` y `www.ESPOL.edu.ec` corresponden al mismo sitio.

Realizar un programa que realice lo siguiente:

- Muestre los nombres o siglas de las universidades que aparecen en la lista (sin repetir).
- Muestre la cantidad y los nombres/siglas de universidades de Ecuador que aparecen en la lista.

##
**Solución Propuesta:**

````python
historial = [
    "www.espol.edu.ec",
    "www.google.com",
    "www.sri.gob.ec",
    "www.fiec.espol.edu.ec",
    "www.uess.edu.ec",
    "www.FIEC.espol.edu.ec",
    "www.fict.espol.edu.ec",
    "www.fcnm.Espol.edu.ec",
    "www.ucsg.edu.ec",
    "www.Stanford.edu",
    "www.harvard.edu",
    "www.stanford.edu",
    "www.UCSG.edu.ec",
    "www.google.com.ec",
    "www.facebook.com",
    "www.opensource.org",
    "www.educacionbc.edu.mx"
]

universidades = []
universidades_ec = []

for url in historial:

    # Convertir a minúsculas para evitar sensibilidad a mayúsculas
    url = url.lower()

    # Verificar si es un sitio educativo
    if ".edu" in url:

        # Dividir la URL en partes para extraer el nombre de la universidad
        partes = url.split(".")

        # Encontrar el índice de ".edu"
        indice = partes.index("edu")

        # El nombre de la universidad está antes de "edu"
        nombre = partes[indice-1].upper()

        # Agregar a la lista de universidades si no está ya presente
        if nombre not in universidades:
                universidades.append(nombre)

        # Verificar si es una universidad ecuatoriana
        if url.endswith(".edu.ec"):
            if nombre not in universidades_ec:
                universidades_ec.append(nombre)

# Mostrar universidades encontradas
print(f"En el historial aparecen {len(universidades)} universidades:")

for i in range(len(universidades)):
    print(f"{i+1}. {universidades[i]}")

# Mostrar universidades ecuatorianas
print("\nUniversidades de Ecuador:")

for i in range(len(universidades_ec)):
    print(f"{i+1}. {universidades_ec[i]}")

# Buscar una URL
buscar = input("\nIngrese una URL: ").lower()

contador = 0

for url in historial:
    if url.lower() == buscar:
        contador += 1

print(f"Este sitio web ha sido visitado {contador} veces")
````
