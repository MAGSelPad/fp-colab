## Ejercicio 3

Una persona realizó una inversión en un banco y desea conocer cuánto dinero ha ganado con el paso del tiempo. El banco aplica una tasa de interés del 12% por cada año completo transcurrido y un 4% por cada mes adicional.

Desarrolle un programa que le pida al usuario que ingrese el monto de su inversión, la fecha inicial, y la fecha final de su inversión. El programa debe calcular y mostrar el tiempo que ha transcurrido entre ambas fechas y la ganancia obtenida por el tiempo de inversión.

**Ejemplo de ejecución:**

```
Ingrese el monto de la inversión: $5000
Ingrese el mes de inicio de la inversión: 10
Ingrese el año de inicio de la inversión: 2019
Ingrese el mes de fin de la inversión: 3
Ingrese el año de fin de la inversión: 2025

Tiempo de inversión transcurrido: 5 año(s) y 5 mes(es)
Ganancia obtenida: $ 4000.0
Monto total acumulado: $ 9000.0
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 3 aquí

````

##
**Solución Propuesta:**

````python
monto = float(input("Ingrese el monto de la inversión: $"))

mes_inicio = int(input("Ingrese el mes de inicio de la inversión: "))
anio_inicio = int(input("Ingrese el año de inicio de la inversión: "))

mes_final = int(input("Ingrese el mes de fin de la inversión: "))
anio_final = int(input("Ingrese el año de fin de la inversión: "))

# Cálculo del tiempo transcurrido

# Conversión a meses para facilitar el cálculo de la diferencia entre fechas.
total_inicio = (anio_inicio * 12) + mes_inicio
total_final = (anio_final * 12) + mes_final

diferencia_meses = total_final - total_inicio

# Cálculo de años y meses a partir de la diferencia en meses
anios = diferencia_meses // 12 # División entera para obtener la cantidad de años completos
meses = diferencia_meses % 12 # Residuo de la división nos da la cantidad de meses restantes ignorando años completos


interes = (anios * 0.12) + (meses * 0.04)
ganancia = monto * interes
total = monto + ganancia

# Resultados
print(f"\nTiempo de inversión transcurrido: {anios} año(s) y {meses} mes(es)")
print("Ganancia obtenida: $", round(ganancia, 2))
print("Monto total acumulado: $", round(total, 2))
````

## Ejercicio 5

En cálculo la derivada de x^4 es 4x^3, la derivada de x^5 es 5x^4. Escriba un programa que permita el ingreso de una ecuación de un solo término y muestre por pantalla la derivada de la misma.

**Nota:** El programa puede implementar soporte **solo para una** de las dos formas: ax^n o x^n

**Ejemplos de ejecución:**

```
Ingrese una ecuación de la forma x^n o ax^n: x^5
La derivada de x^5 es 5x^4
```

```
Ingrese una ecuación de la forma x^n o ax^n: 10y^15
La derivada de 10y^15 es 150y^14
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 5 aquí

````

##
**Solución Propuesta:**

````python
#Se solicita ingreso de la ecuación
ecuacion = input("Ingrese una ecuación de la forma x^n o ax^n: ")

# Se separa la base del exponente gracias al '^' en 2 variables distintas
base, exponente = ecuacion.split("^")

# Se obtiene la variable, que es el último caracter antes del '^'
variable = base[-1]

# Se obtiene el coeficiente, que es la parte de la base antes de la variable. Si no hay coeficiente, se asume que es 1.
coeficiente = ecuacion.split(variable)[0] or "1"

derivada = f"{int(coeficiente) * int(exponente)}{variable}^{int(exponente)-1}"
print(f"La derivada de {ecuacion} es {derivada}")
````

## Ejercicio 7

Escriba un programa que pida por teclado los segundos (más de 3600) y muestre por pantalla la hora en formato *hh:mm:ss*

**Ejemplo de ejecución:**

```
Ingrese una cantidad de segundos: 36923
36923 segundos en formato hh:mm:ss es igual a: 10:15:23
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 7 aquí

````

##
**Solución Propuesta:**

````python
segundos_total = int(input("Ingrese una cantidad de segundos: "))

# Conversión de segundos a horas y nos quedamos solo con la parte entera.
horas = segundos_total // 3600

# El residuo de la conversión de seg a h es igual a la cantidad de minutos en segundos
# Convertimos estos segundos a minutos, nos quedamos solo con la parte entera.
minutos = (segundos_total % 3600) // 60

# El residuo de la conversión de minutos a segundos son los segundos restantes.
segundos = (segundos_total % 3600) % 60

tiempo = "%02d:%02d:%02d" % (horas, minutos, segundos)

print(f"{segundos_total} segundos en formato hh:mm:ss es igual a: {tiempo}")
````

Desarrolle un juego que mezcle palabras.

Para empezar se debe crear un programa que mezcle dos palabras de la siguiente manera:

- Para cada palabra se debe elegir una letra al azar,
- a partir de esta letra, junto a las siguientes se utilizan para la mezcla.

**Ejemplo de ejecución:**

```
Ingrese palabra 1: maletas
Letra al azar: e
Resultado 1: etas
Ingrese palabra 2: pesadisimas
Letra al azar: d
Resultado 2: disimas
Mezcla: etasdisimas
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 8 aquí

````

##
**Solución Propuesta:**

````python
import random


palabra1 = input("Ingrese palabra 1: ")

# Generamos un índice aleatorio dentro del rango de la palabra
# len(palabra1)-1 es el último índice válido
indice1 = random.randint(0, len(palabra1)-1)

# Obtenemos la letra que está en ese índice aleatorio
letra1 = palabra1[indice1]
print("Letra al azar:", letra1)

# Tomamos desde la letra aleatoria hasta el final de la palabra
resultado1 = palabra1[indice1:]
print("Resultado 1:", resultado1)



palabra2 = input("Ingrese palabra 2: ")

# Generamos otro índice aleatorio para la segunda palabra
indice2 = random.randint(0, len(palabra2)-1)

# Obtenemos la letra en esa posición
letra2 = palabra2[indice2]
print("Letra al azar:", letra2)

# Tomamos desde esa letra hasta el final de la palabra
resultado2 = palabra2[indice2:]
print("Resultado 2:", resultado2)



mezcla = resultado1 + resultado2
print ("Mezcla:", mezcla)
````

A usted se le presenta el siguiente string:

`datos = "<div><name>Pamela Nayeli</name><apellido>Rugel Díaz</apellido><edad>20<edad></div>"`

Se le solicita obtener la información del string y mostrarla como en el siguiente

**Ejemplo de ejecución:**

```
Nombres: Pamela Nayeli
Apellidos: Rugel Díaz
Edad: 20
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 10 aquí

````

##
**Solución Propuesta:**

````python
datos = "<div><name>Pamela Nayeli</name><apellido>Rugel Díaz</apellido><edad>20</edad></div>"

# Extraer nombres

# Índice donde empieza "<name>" + len para saltarnos la etiqueta
inicio_nombre = datos.find("<name>") + len("<name>")

# Índice donde empieza "</name>"
fin_nombre = datos.find("</name>")

# Nos interesa lo que hay entre ambos índices
nombres = datos[inicio_nombre : fin_nombre]


# Extraer apellidos
inicio_apellido = datos.find("<apellido>") + len("<apellido>")
fin_apellido = datos.find("</apellido>")
apellidos = datos[inicio_apellido : fin_apellido]


# Extraer edad
inicio_edad = datos.find("<edad>") + len("<edad>")
fin_edad = datos.find("</edad>")
edad = datos[inicio_edad : fin_edad]


print("Nombres:", nombres)
print("Apellidos:", apellidos)
print("Edad:", edad)
````

UESS
ESPOL

##

````python
url1 = input("Ingrese dirección de universidad 1: ").lower()
url2 = input("Ingrese dirección de universidad 2: ").lower()


# Encontramos índice de inicio después de "www."
inicio1 = url1.find("www.") + 4 # +4 para saltarnos "www."

# Encontramos índice dónde empieza ".edu"
fin1 = url1.find(".edu")

# Extraemos lo que está entre ambos índices
parte1 = url1[inicio1 : fin1]

# Nos quedamos con la parte separada por punto
# Si no hay punto, find(".") retorna -1, por lo que se toma desde el inicio ([0:])
sigla1 = parte1[parte1.find(".") + 1 : ].upper()


# Repetimos el proceso para la segunda URL
inicio2 = url2.find("www.") + 4
fin2 = url2.find(".edu")
parte2 = url2[inicio2 : fin2]
sigla2 = parte2[parte2.find(".") + 1 : ].upper()


print(sigla1)
print(sigla2)
````

##
**Solución Propuesta (Utilizando split):**

````python
url1 = input("Ingrese dirección de universidad 1: ").lower()
url2 = input("Ingrese dirección de universidad 2: ").lower()


# Separamos la URL en dos partes usando "www."
# Ejemplo: www.fiec.espol.edu.ec → ["", "fiec.espol.edu.ec"]
parte1 = url1.split("www.")[1] # Tomamos elemento de índice 1

# Ahora separamos antes de ".edu"
# Ejemplo: "fiec.espol.edu.ec" → ["fiec.espol", "ec"]
antes_edu1 = parte1.split(".edu")[0]  # Tomamos elemento de índice 0

# Tomamos la última parte después del punto
# Ejemplo: "fiec.espol" → ["fiec", "espol"] → "espol"
# Si no hay punto, se toma la única parte que hay
sigla1 = antes_edu1.split(".")[-1].upper()


# Repetimos el proceso para la segunda URL
parte2 = url2.split("www.")[1]
antes_edu2 = parte2.split(".edu")[0]
sigla2 = antes_edu2.split(".")[-1].upper()


print(sigla1)
print(sigla2)
````

````python
import random

# -------------------- 1. INGRESO DE DATOS --------------------

# Datos personales
nombre_completo = input("Ingrese nombre completo: ")
fecha_nacimiento = input("Ingrese fecha de nacimiento (dd-mm-aaaa): ")
matricula = input("Ingrese número de matrícula: ")
correo = input("Ingrese correo electrónico: ")

# Materias y notas
materia1 = input("Ingrese materia 1: ")
nota1 = float(input("Ingrese nota de la materia 1: "))

materia2 = input("Ingrese materia 2: ")
nota2 = float(input("Ingrese nota de la materia 2: "))

materia3 = input("Ingrese materia 3: ")
nota3 = float(input("Ingrese nota de la materia 3: "))

# -------------------- 2. LISTA DE NOMBRE + USUARIO --------------------

# Separar nombres y apellidos
partes_nombre = nombre_completo.split()

# Usuario del correo (antes del @)
usuario = correo.split("@")[0]

# Lista final
lista_datos = partes_nombre + [usuario]

# -------------------- 3. CÁLCULO DE EDAD --------------------
dia_nac, mes_nac, anio_nac = fecha_nacimiento.split("-")
anio_nac = int(anio_nac)
mes_nac = int(mes_nac)

anio_act, mes_act = 2026, 5 # Constantes con la fecha actual

edad = anio_act - anio_nac
edad = edad - (mes_act < mes_nac)  # (True = 1, False = 0)

# -------------------- 4. AÑO DE INGRESO --------------------
# La matrícula contiene el año al inicio (ej: 202312345)
anio_ingreso = matricula[:4]

# -------------------- 5. PROMEDIO --------------------
promedio = (nota1 + nota2 + nota3) / 3

# -------------------- 6. APROBACIÓN --------------------
estado = promedio >= 60

# -------------------- 7. TURNO ALEATORIO --------------------
turno = random.randint(100, 999)

# -------------------- 8. REPORTE FINAL --------------------
print("\nREPORTE DEL ESTUDIANTE:")
print("Nombres y usuario:", lista_datos)
print("Edad:", edad, "años")
print("Año de ingreso:", anio_ingreso)
print(f"Promedio: {promedio:.2f}")
print("Aprobado?", estado)
print("Turno de registro:", turno)
````
