Una estudiante quiere conocer si dispone del dinero suficiente para:

- Pagar una suscripción de Spotify,
- comprar una skin en su juego favorito,
- hacer un pedido de delivery.

Escriba un programa que le indique si puede hacer los gastos y que calcule cuánto dinero le queda después de realizar los gastos.

- Solicite el dinero que dispone la estudiante.
- Defina las siguientes variables:
    - `spotify = 3.99`
    - `skin = 16.99`
    - `delivery = 12`

**Ejemplo de ejecución:**

```
Ingrese su dinero disponible: 50
```

## Ejercicio 3

Escriba un programa que solicite al usuario ingresar 5 nombres de estudiantes separados por comas (”,”). El programa debe elegir un nombre al azar y mostrar quien fue seleccionado. Además, debe mostrar cuántos caracteres tiene el nombre seleccionado.

**Ejemplo de ejecución:**

```
Ingrese 5 nombres separados por comas: Ana, Pedro, María, Luis, Carmen
Nombre sorteado: Ana
El nombre tiene 3 caracteres.
```

**Solución Propuesta (Utilizando índices)**

````python
import random

entrada = input("Ingrese 5 nombres separados por comas: ")

# Se realiza una extracción manual de los nombres usando índices
nombre1 = entrada[:entrada.find(",")]
resto1 = entrada[entrada.find(",")+1 : ]

nombre2 = resto1[:resto1.find(",")]
resto2 = resto1[resto1.find(",")+1 : ]

nombre3 = resto2[:resto2.find(",")]
resto3 = resto2[resto2.find(",")+1 : ]

nombre4  = resto3[:resto3.find(",")]
nombre5 = resto3[resto3.find(",")+1:]

# Elegir al azar
sorteado = random.choice([nombre1, nombre2, nombre3, nombre4, nombre5])
sorteado_limpio = sorteado.strip()
# strip() para eliminar espacios en blanco al inicio o al final del nombre

print("Nombre sorteado:", sorteado_limpio)
print("El nombre tiene", len(sorteado_limpio), "caracteres")
````

## Ejercicio 4

Simular el juego de la lotería con 5 tómbolas que generen números aleatorios del 1 al 9 para formar el número ganador de la primera suerte.

El programa deberá solicitar al usuario el ingreso de un número e identificar si se lleva el premio mayor, o si al menos gana reintegro.

Para ganar el premio mayor deben coincidir los 5 números, mientras que para ganar reintegro sólo debe coincidir el último número.

Sólo puede ganar un premio de acuerdo a la relevancia del mismo.

**Ejemplos de ejecución:**

```
Ingrese su número: 12345
El número ganador es: 48722
¿Usted ha sido ganador del premio mayor? False
¿Usted ha sido ganador del reintegro? False
¿Su número no obtuvo premio? True
```

```
Ingrese su número: 12345
El número ganador es: 95445
¿Usted ha sido ganador del premio mayor? False
¿Usted ha sido ganador del reintegro? True
¿Su número no obtuvo premio? False
```

```
Ingrese su número: 24851
El número ganador es: 24851
¿Usted ha sido ganador del premio mayor? True
¿Usted ha sido ganador del reintegro? False
¿Su número no obtuvo premio? False
```

**Solución Propuesta**

````python
import random

# Entrada del usuario
numero_usuario = input("Ingrese su número: ")

# Generación de 5 números aleatorios (tómbolas)
n1 = str(random.randint(1, 9))
n2 = str(random.randint(1, 9))
n3 = str(random.randint(1, 9))
n4 = str(random.randint(1, 9))
n5 = str(random.randint(1, 9))

numero_ganador = n1 + n2 + n3 + n4 + n5

print("El número ganador es:", numero_ganador)

# Verificación del premio mayor
premio_mayor = (numero_usuario == numero_ganador)

# Verificación de reintegro
reintegro = (numero_usuario[-1] == numero_ganador[-1])

# Mostrar resultados
print("¿Usted ha sido ganador del premio mayor?", premio_mayor)
print("¿Usted ha sido ganador del reintegro?", reintegro and not premio_mayor)
print("¿Su número no obtuvo premio?", not premio_mayor and not reintegro)
````

## Ejercicio 5

Se dispone del siguiente código que permite generar las letras del abecedario.

```python
import string

abc=string.ascii_lowercase
print(abc)
```

Se solicita realizar un programa que genere un string que contenga las letras del abecedario (ver ejemplo del código anterior) y finalmente genere una palabra que contenga entre 6 a 10 letras del abecedario.

Las letras no se pueden repetir en la palabra, y una de esas letras debe estar en mayúscula y es elegida aleatoriamente.

No es necesario que la palabra contenga vocales.

**Nota:** Considere utilizar el código proporcionado en la ubicación más adecuada de su código.

**Ejemplos de ejecución:**

```
Abecedario: abcdefghijklmnopqrstuvwxyz
Palabra generada: sxAnuq
```

```
Abecedario: abcdefghijklmnopqrstuvwxyz
Palabra generada: Dcvxlhyujr
```

```
Abecedario: abcdefghijklmnopqrstuvwxyz
Palabra generada: rladbJm
```

**Solución Propuesta**

````python
import string
import random

# Generar letras del abecedario
abc = string.ascii_lowercase
print("Abecedario:", abc)

# Cantidad aleatoria de letras entre 6 y 10
cantidad = random.randint(6, 10)

# Seleccionar letras sin repetir
letras = random.sample(abc, cantidad)

# Elegir una posición aleatoria para convertir en mayúscula
pos_mayuscula = random.randint(0, cantidad - 1)
letras[pos_mayuscula] = letras[pos_mayuscula].upper()

# Unir las letras en una palabra
palabra = "".join(letras)

print("Palabra generada:", palabra)
````
