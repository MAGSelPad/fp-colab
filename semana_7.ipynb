## Ejercicio 1

Implemente la función `buscar_palindromos(mensaje)` que recibe un mensaje y retorna una lista con todas las palabras (de 2 o más letras) del mensaje que son palíndromos. Asuma que todas las palabras están en minúsculas y no existen signos de puntuación.

**Recuerde que un palíndromo es una cadena que se lee igual de izquierda a derecha o de derecha a izquierda. Por ejemplo:** madam, ana, somos, reconocer, anilina.

**Ejemplo de ejecución:**

Con la entrada:

```python
mensaje = "ana y yo somos amigos y trabajamos en la torre del radar"
```

La función retorna:

```
['ana', 'somos', 'radar']
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 1 aquí

````

##
**Solución Propuesta:**

````python
def buscar_palindromos(mensaje):
    palindromos = []
    palabras = mensaje.split()

    for palabra in palabras:
        if len(palabra) >= 2 and palabra == palabra[::-1]:
            palindromos.append(palabra)

    return palindromos

mensaje = "ana y yo somos amigos y trabajamos en la torre del radar"
print(buscar_palindromos(mensaje))
````

## Ejercicio 2

Implemente la función `siglas(frase, palabras_comunes)` que reciba una frase y una lista de palabras comunes. Asuma que tanto las palabras de la frase como de la lista están en minúsculas. La función debe generar una sigla utilizando las iniciales en mayúsculas de cada palabra. La función debe omitir de la sigla palabras comunes como "de", "y", "en", etc. (definidas en el parámetro `palabras_comunes`), a menos que sean la primera o la última palabra de la frase. Asuma que en la frase no existen signos de puntuación.

**Ejemplo de ejecución:**

Con la entrada:

```python
frase = "el yin y el yang son conceptos filosóficos de la cultura china"
palabras_comunes = ["el", "y", "de", "la", "son"]
```

La función retorna:

```
EYYCFCC
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 2 aquí

````

##
**Solución Propuesta:**

````python
def siglas(frase, palabras_comunes):
    palabras = frase.split()
    sigla = ""

    for i in range(len(palabras)):
        palabra = palabras[i]
        if i == 0 or i == len(palabras) - 1:
            sigla += palabra[0].upper()
        elif palabra not in palabras_comunes:
            sigla += palabra[0].upper()

    return sigla

frase = "el yin y el yang son conceptos filosóficos de la cultura china"
palabras_comunes = ["el", "y", "de", "la", "son"]
print(siglas(frase, palabras_comunes))
````

## Ejercicio 3
Implemente la función `reemplazar_palabras(mensaje, palabras)` que recibe un mensaje y una lista de strings donde cada string tiene el formato **"palabra_vieja|palabra_nueva"**. Para cada string en la lista, la función debe reemplazar todas las ocurrencias exactas de la palabra vieja por la palabra nueva.

La función retorna el nuevo mensaje.

**Asuma que todas las palabras en el mensaje están en minúsculas, separadas por un espacio en blanco y que el mensaje no contiene signos de puntuación.**

Note que en el ejemplo, la palabra "lleno" en "relleno" no se reemplaza por "repleto" por no ser una ocurrencia exacta.

**Ejemplo de ejecución:**

Con la entrada:

```python
mensaje = "comer relleno me deja muy lleno y cuando estoy lleno no me siento bien"
palabras = ["lleno|repleto", "comer|cenar", "bien|alegre"]
```

La función retorna:

```
cenar relleno me deja muy repleto y cuando estoy repleto no me siento alegre
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 3 aquí

````

##
**Solución Propuesta:**

````python
def reemplazar_palabras(mensaje, palabras):
    lista_palabras = mensaje.split()

    for reemplazo in palabras:
        vieja, nueva = reemplazo.split("|")

        for i in range(len(lista_palabras)):
            if lista_palabras[i] == vieja:
                lista_palabras[i] = nueva

    return " ".join(lista_palabras)

mensaje = "comer relleno me deja muy lleno y cuando estoy lleno no me siento bien"
palabras = ["lleno|repleto", "comer|cenar", "bien|alegre"]
print(reemplazar_palabras(mensaje, palabras))
````

## Ejercicio 4

Implemente un programa que simule una carrera de 10 turnos entre dos animales: `Liebre` y `Tortuga`. Cada animal tiene una velocidad máxima guardada en `vel_l` y `vel_t` respectivamente, que indica cuánto puede avanzar por turno, usando un número aleatorio entre 1 y Ia velocidad máxima, ambos incluidos.

Además, hay una lista llamada `l_trampas` que indica las casillas donde hay trampas para la liebre. Si cae en una, regresa a la posición anterior (donde estaba antes del último turno).

El avance de cada animal se acumula turno a turno, sumando las casillas recorridas en cada movimiento para determinar su posición final en la pista de la carrera. Además, debe mostrar con guiones el avance alcanzado como se muestra en el ejemplo de salida esperada.

Al final de los 10 turnos, gana el animal que haya avanzado más casillas en total. En caso de empate, se declara empate.

**Ejemplo de ejecución:**

Con la entrada:

```python
# Datos iniciales:
vel_l = 5
vel_t = 3
l_trampas = [4, 7, 9, 13]
```

Se imprime el resultado:

```
Turno 1:
Liebre: -- (posición 2)
Tortuga: -- (posición 2)

Turno 2:
Liebre: --- (posición 3)
Tortuga: --- (posición 3)

Turno 3:
Liebre cayó en trampa en casilla 4, regresa a su posición anterior
Liebre: --- (posición 3)
Tortuga: ------ (posición 6)

Turno 4:
Liebre: ------ (posición 6)
Tortuga: -------- (posición 8)

...

Turno 10:
Liebre: ------------------------------ (posición 30)
Tortuga: ----------------------- (posición 23)

Ganador: Liebre
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 4 aquí

````

##
**Solución Propuesta:**

````python
import random

# Datos iniciales:
vel_l = 5
vel_t = 3
l_trampas = [4, 7, 9, 13]
pos_liebre = 0
pos_tortuga = 0

for turno in range(1, 11):
    print(f"Turno {turno}:")

    pos_anterior = pos_liebre

    avance_l = random.randint(1, vel_l)
    avance_t = random.randint(1, vel_t)
    pos_liebre += avance_l
    pos_tortuga += avance_t

    # Verificar trampas
    if pos_liebre in l_trampas:
        print(f"Liebre cayó en trampa en casilla {pos_liebre}, regresa a su posición anterior")
        pos_liebre = pos_anterior

    # Mostrar pista recorrida
    print(f"Liebre: {'-' * pos_liebre} (posición {pos_liebre})")
    print(f"Tortuga: {'-' * pos_tortuga} (posición {pos_tortuga})")
    print()

# Determinar ganador
if pos_liebre > pos_tortuga:
    print("Ganador: Liebre")
elif pos_tortuga > pos_liebre:
    print("Ganador: Tortuga")
else:
    print("Empate")
````
