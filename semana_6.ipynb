## Ejercicio 1

Crea una función `primera_letra_mayuscula(frase, palabras_excluidas)` que reciba una frase y una lista de palabras comunes que deben ser excluidas de la sigla. La función debe devolver la sigla formada por las primeras letras de las palabras que no están en la lista `palabras_excluidas`, en mayúsculas.

**Instrucciones:**

- Llama a la función con los valores `8` y `2` y muestra los resultados en la consola.

**Ejemplo de ejecución:**
Dada las entradas:
```
frase="el libro de la vida es fascinante y hermoso"
palabras_excluidas= ["el","de","la","y"]
```
Se debe mostrar:
```
LVEFH
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 1 aquí

````

##
**Solución Propuesta:**

````python
def primera_letra_mayuscula(frase, palabras_excluidas):
    sigla = ""
    palabras = frase.split()
    for palabra in palabras:
        if palabra not in palabras_excluidas:
            sigla += palabra[0].upper()
    return sigla


# Programa principal
frase = "el libro de la vida es fascinante y hermoso"
palabras_excluidas = ["el", "de", "la", "y"]

print(primera_letra_mayuscula(frase, palabras_excluidas))
````

Crea dos funciones:

- `centimetros_a_milimetros(centimetros)` que convierta una longitud en centímetros a milímetros.

Luego, crea una **tercera función** llamada `metros_a_milimetros(metros)` que **llame** a las dos funciones anteriores para convertir directamente de metros a milímetros.

````python
# Desarrolle solución del EJERCICIO 2 aquí

````

## Ejercicio 7

Define una función `rotacion_izquierda(numeros, k)` que reciba una lista de números y un número entero `k`. La función debe rotar hacia la izquierda los primeros `k` elementos de la lista y devolver la nueva lista.

**Ejemplo de ejecución:**
Dada la entrada:
```
numeros= [1,2,3,4,5,6,7]
k=3
```
Se debe mostrar:
```
[2,3,1,4,5,6,7]
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 7 aquí

````

##
**Solución Propuesta:**

````python
def rotacion_izquierda(numeros, k):
    primeros = numeros[:k] #Primer grupo de elementos, antes de k
    elementos2 = numeros[k:] #Segundo grupo de elementos, despues de k
    primer_primeros = primeros[0] #Primer elemento del primer grupo
    restantes_elemento1 = primeros[1:] #Elementos que restan del primer grupo
    primeros_modificada = restantes_elemento1 + [primer_primeros] #Lista con la rotación realizada a la izquierda en el primer grupo
    return primeros_modificada + elementos2


# Programa principal
numeros = [1, 2, 3, 4, 5, 6, 7]
print(rotacion_izquierda(numeros, 3))
````

```
empleados= [
```

### 1. Crear la función `filtrar_por_estado`

La función recibe:

- `l_empleados`: lista de empleados con el formato `"Nombre-Estado-Departamento"`
- `estado`: cadena de texto (`"activo"` o `"inactivo"`)

La función debe devolver una lista con todos los empleados que coincidan con el estado solicitado.

**Ejemplo de ejecución:**

Si:

```
filtrar_por_estado(empleados,"activo")
```

debe retornar:

```
"Luisa-activo-Finanzas", "Sofia-activo-Marketing", "Daniel-activo-Ventas", "Elena-activo-RRHH"]
```


### 2. Crear la función `empleados_destacados`

La función recibe:

- `l_empleados`: lista de empleados con el formato `"Nombre-Estado-Departamento"`
- `l_indicadores`: lista de indicadores con el formato `"Nombre-Valor-TipoIndicador"`
- `valor_minimo`: entero que representa el valor mínimo requerido para el indicador `"proyectos"`. Su valor predeterminado es: 100.

La función debe devolver una lista con los empleados destacados, es decir, aquellos que:

1. Se encuentren **activos**.
2. Tengan un indicador de tipo `"proyectos"`.
3. Posean una cantidad de proyectos mayor al `valor_minimo`.

**Ejemplo de ejecución:**

```
["Pedro-120-proyectos", "Sofia-110-proyectos", "Daniel-130-proyectos"]
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 1 aquí

````

##
**Solución Propuesta:**

````python
def filtrar_por_estado(l_empleados, estado):
    resultado = []

    for empleado in l_empleados:
        datos = empleado.split("-")
        if datos[1] == estado:
            resultado.append(empleado)

    return resultado

def empleados_destacados(l_empleados, l_indicadores, valor_minimo=100):
    destacados = []
    activos = filtrar_por_estado(l_empleados, "activo")

    for indicador in l_indicadores:
        datos = indicador.split("-")
        nombre = datos[0]
        valor = int(datos[1])
        tipo = datos[2]
        if tipo == "proyectos" and valor > valor_minimo:
            for empleado in activos:
                datos_empleado = empleado.split("-")
                if datos_empleado[0] == nombre:
                    destacados.append(indicador)

    return destacados


# Programa principal
empleados= [
"Ana-activo-Ventas",
"Carlos-inactivo-Marketing",
"Pedro-activo-Sistemas",
"Maria-activo-RRHH",
"Jorge-inactivo-Logistica",
"Luisa-activo-Finanzas",
"Sofia-activo-Marketing",
"Miguel-inactivo-Sistemas",
"Daniel-activo-Ventas",
"Elena-activo-RRHH"
]

indicadores= [
"Ana-95-desempeno",
"Carlos-80-desempeno",
"Pedro-120-proyectos",
"Maria-70-desempeno",
"Jorge-150-proyectos",
"Luisa-85-desempeno",
"Sofia-110-proyectos",
"Miguel-60-desempeno",
"Daniel-130-proyectos",
"Elena-90-desempeno"
]

print(empleados_destacados(empleados, indicadores))
````

## Ejercicio 2

Un nutricionista está implementando un nuevo plan nutricional con sus pacientes. Según las condiciones individuales de cada uno establece metas personalizadas de pérdida de eso semanal que deben alcanzar.

### Cree la función `verificar_peso` que recibe:

- `l_pesos`: una lista que contiene los registros de peso semanal de una persona.
- `libras`: la cantidad minima de libras que la persona debe perder cada semana para cumplir con el objetivo.

La función devuelve dos valores:

- La cantidad de semanas que NO se alcanzó el objetivo, es decir, perdió menos libras que la cantidad especificada.
- La racha máxima, es decir, la cantidad máxima de semanas consecutivas en las que se alcanzó el objetivo de pérdida de peso.

**Ejemplo de ejecución:**

Si:

```python
libras = 2
pesos = [200, 198, 196, 195, 193, 192, 189]

incumplidas, racha = verificar_peso(pesos, libras)

print("Semanas sin cumplir =", incumplidas)
print("Racha Máxima =", racha)
```

Al ejecutarse se muestra:

```
Semanas sin cumplir = 2
Racha Máxima = 2
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 2 aquí

````

##
**Solución Propuesta:**

````python
def verificar_peso(l_pesos, libras):
    semanas_incumplidas = 0
    racha_actual = 0
    racha_maxima = 0

    for i in range(1, len(l_pesos)):
        perdida = l_pesos[i - 1] - l_pesos[i]
        if perdida >= libras:
            racha_actual += 1
            if racha_actual > racha_maxima:
                racha_maxima = racha_actual
        else:
            semanas_incumplidas += 1
            racha_actual = 0

    return semanas_incumplidas, racha_maxima

# Programa principal
libras = 2
pesos = [200, 198, 196, 195, 193, 192, 189]

incumplidas, racha = verificar_peso(pesos, libras)

print("Semanas sin cumplir =", incumplidas)
print("Racha Máxima =", racha)
````

## Ejercicio 3

Implemente la función `validar(cadena)` que recibe un string.

La función debe retornar `True` si la cadena cumple con todos los siguientes requisitos o `False` en caso contrario.

Requisitos a validar:

1. Tener al menos 8 caracteres
2. Incluir al menos un letra mayúscula
3. Incluir al menos un número
4. Incluir al menos uno de estos caracteres especiales (#, $, ^)

Luego, en su programa principal, asuma que tiene una lista con varias cadenas. Utilice la función que creó para validar las cadenas y mostrar por pantalla cuantas de ellas cumplen con los requisitos.

**Ejemplo de ejecución:**

Si:

```python
cadenas = ["Hola123#", "python123", "ESPOL2025$", "ClaveSegura9^", "abc", "Programacion1"]

contador = 0

# Lógica de conteo de contraseñas válidas
# ...

print("Cantidad de cadenas válidas:", contador)
```

Al ejecutarse se muestra:

```
Cantidad de cadenas válidas: 3
```

**Desarrolle:**

````python
# Desarrolle solución del EJERCICIO 3 aquí

````

##
**Solución Propuesta:**

````python
def validar(cadena):
    tiene_mayuscula = False
    tiene_numero = False
    tiene_especial = False
    for caracter in cadena:
        if caracter.isupper():
            tiene_mayuscula = True
        if caracter.isdigit():
            tiene_numero = True
        if caracter in "#$^":
            tiene_especial = True
    return len(cadena) >= 8 and tiene_mayuscula and tiene_numero and tiene_especial


# Programa principal

cadenas = ["Hola123#", "python123", "ESPOL2025$", "ClaveSegura9^", "abc", "Programacion1"]
contador = 0

# Lógica de conteo
for cadena in cadenas:
    if validar(cadena):
        contador += 1

print("Cantidad de cadenas válidas:", contador)
````
